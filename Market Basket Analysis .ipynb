```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('supermarket.csv')
```

```python
df.head(10)
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df1=df["MILK,BREAD,BISCUIT"].apply(lambda t: t.split(","))
```

```python
df1
```

```python
df1=list(df1)
df1
```

```python
te=TransactionEncoder()
```

```python
df2=te.fit(df1).transform(df1)
```

```python
fte=pd.DataFrame(df2,columns=te.columns_)
```

```python
fte
```

```python
freq=apriori(fte,min_support=0.1,use_colnames=True)
```

```python
freq
```

```python
asr=association_rules(freq,metric="confidence",min_threshold=0.3)
```

```python
asr
```

```python
asr=association_rules(freq,metric="lift",min_threshold=0.3)
```
